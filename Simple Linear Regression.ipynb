# Implementing a simple Linear Regression Model  
### The model $f_{w,b}$ for linear regression with one variable

## STEP 1: Import Relevant Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## STEP 2: Load the data 

```python
x_input = np.array([1.0, 2.0])
y_output = np.array([300, 500])
print(x_input)
print(y_output)
```

```python
x_input.shape
```

```python
y_output.shape
```

```python
# num of training examples
m = x_input.shape[0]
print(m)
m = len(x_input)
print(m)
```

## STEP 3: Set the $^{(i)}$th Training Example

```python
# initialize i to 0
i = 0
x_i = x_input[i]
y_i = y_output[i]
print(i)
print(x_i)
print(y_i)
```

## STEP 4: Visualize the data 

```python
plt.scatter(x_input, y_output, marker = 'x', c='r')
plt.title('House Price Predictor')
plt.ylabel('Price in 1000s')
plt.xlabel('Size in sqft')
plt.show()
```

## STEP 5: Build the Model 

```python
# define the parameters 
weight = 200
bias = 100

# function 
def house_price_predictor(data, weight, bias):
    data_size = data.shape[0]
    # initialize prediction array
    prediction = np.zeros(data_size)
    for i in range(data_size):
        prediction[i] = weight * data[i] + bias
    return prediction
    
```

## STEP 6: Call the function

```python
model_output = house_price_predictor(x_input, weight, bias)
model_output
```

## STEP 7: Visualize Prediction

```python
# Plot of model prediction
plt.plot(x_input, model_output, c='r', marker='.',label='Prediction')

# plot of our model 
plt.scatter(x_input, y_output, c='b', marker='x', label='Actual')

plt.title('Price Prediction')
plt.xlabel('Size in sqft')
plt.ylabel('Price in 1,000s')
plt.show()
```
